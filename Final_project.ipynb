```python
tips_df.dropna(subset = ["Surname","Geography","Tenure","Balance","NumOfProducts","HasCrCard","IsActiveMember","Exited"], inplace=True, how='any')
tips_df
```

```python
tips_df['Gender'].fillna(tips_df['Gender'].mode()[0], inplace = True)
tips_df['EstimatedSalary'].fillna(tips_df['EstimatedSalary'].mean , inplace = True)
tips_df
```

```python
tips_df.isnull().any()
```

```python
tips_df.info()
```

```python
Geography_set=["3","4",'10000','56','Pooya']
for i in range(len(Geography_set)):
    index_Geography = tips_df.index[tips_df.Geography == Geography_set[i]]
    tips_df.drop(index=index_Geography, axis=0, inplace=True)
    
index_Age = tips_df.index[tips_df.Age == -10]
tips_df.drop(index=index_Age, axis=0, inplace=True)


index_HasCrCard = tips_df.index[tips_df.HasCrCard == -1000]
tips_df.drop(index=index_HasCrCard, axis=0, inplace=True)

index_IsActiveMember = tips_df.index[tips_df.IsActiveMember == -1]
tips_df.drop(index=index_IsActiveMember, axis=0, inplace=True)

tips_df
```

```python
Gender_tips_mode=tips_df['Gender'].mode()[0]
tips_df['Gender'].replace('9', Gender_tips_mode, inplace=True)
tips_df['Gender'].replace('5', Gender_tips_mode, inplace=True)
tips_df
```
